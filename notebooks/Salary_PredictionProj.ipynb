```python
# ğŸ“¦ Import necessary libraries
import pandas as pd                         # for reading CSV and handling data
from sklearn.linear_model import LinearRegression  # for creating the prediction model
from sklearn.model_selection import train_test_split  # for splitting the data
from sklearn.metrics import mean_squared_error, r2_score  # for evaluating the model

# ğŸ“‚ Step 1: Load the dataset
df = pd.read_csv("../data/salary_prediction.csv")   # read CSV file
print(df.head())  # show the top rows of data

# ğŸ§  Step 2: Split into features (X) and target (y)
X = df[['Experience']]  # input feature (must be 2D)
y = df['Salary']        # target/output value

# ğŸ”€ Step 3: Split data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# ğŸ—ï¸ Step 4: Create and train the linear regression model
model = LinearRegression()     # create the model
model.fit(X_train, y_train)    # train the model using training data

# ğŸ”® Step 5: Make predictions on test data
y_pred = model.predict(X_test)  # predict salaries for test experiences

# ğŸ“Š Step 6: Print predictions with actual values
results = pd.DataFrame({'Experience': X_test['Experience'], 'Actual Salary': y_test, 'Predicted Salary': y_pred})
print(results)

# ğŸ“ˆ Step 7: Evaluate the model
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))  # how wrong is the prediction
print("RÂ² Score:", r2_score(y_test, y_pred))  # how good is the model (1.0 = perfect)

# ğŸ§ª Step 8: Predict salary for someone with 6.5 years experience
exp = pd.DataFrame([[6.5]], columns=['Experience'])  # create DataFrame with proper feature names
predicted_salary = model.predict(exp)
print(f"Predicted salary for 6.5 years of experience: ${predicted_salary[0]:.2f}")
```
